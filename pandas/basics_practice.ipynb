# Mathematical Functions & Indexing in Series

```python
a = [1, 2, 3]
b = [10, 20, 30]
```

```python
s1 = pd.Series(
    [1, 2, 3],
    index=["a", "b", "c"]
)

s2 = pd.Series(
    [10, 20, 30],
    index=["a", "b", "c"]
)

print(s1 + s2)
```

```python
s1 = pd.Series(
    [5, 10],
    index=["x", "y"]
)

s2 = pd.Series(
    [100, 200],
    index=["y", "x"]
)

print(s1 + s2)
```

```python
s1 = pd.Series(
    [5, 10, 15],
    index=["x", "y", "z"]
)

s2 = pd.Series(
    [100, 200, 300],
    index=["y", "x", "a"]
)

print(s1.add(s2, fill_value = 0))
s1.add(100)
```

```python
s = pd.Series([4, 7, 2, 9, 1, 5])
print(s.sum(),
s.min(),
s.max(),
s.mean())
```

## loc & ilok
#### slicing in series

```python
s = pd.Series(
    [10, 20, 30, 40, 50],
    index=["a", "b", "c", "d", "e"]
)
```

### loc uses labels and both starting and ending labels are included

```python
s.loc["b":"d"]
```

### iloc uses index numbers/positions like pyhton list slicing and ending position is excluded

```python
s.iloc[1:4]
```
