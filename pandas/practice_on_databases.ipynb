## DataFrames Attributes


```python
df = pd.DataFrame({
    "Name": ["Rahul", "Priya", "Aman"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 88]
})
```

```python
df.shape
```

```python
df.columns
```

```python
df.index
```

```python
df.dtypes
```

```python
df.head(2) # head() returns rows from the top instead of full dataframe
```

```python
df.tail(1) # tail() returns rows from the bottom 
```

```python
df.head(100) # if the input value is more than rows than it will return the available number of rows same for tail()
```

```python
df.sample(2) # sample() returns random rows 
```

```python
df.describe() 

#df.describe()
# Answers different questions:

# What's the average?
# What's the minimum?
# What's the maximum?
# What's the standard deviation?
# What are the quartiles (25%, 50%, 75%)?

# It's the statistical summary of your numeric columns
```

```python
df.info()

# df.info()

# Answers questions like:

# How many rows are there?
# How many columns are there?
# What are the column names?
# What is the data type of each column?
# How many non-null (non-missing) values does each column have?
# How much memory is being used?

# It's basically the identity card of your DataFrame.
```

```python
df.rename(columns={"Marks": "Score"}) # Returns a nnew data frame with new column name
```

```python
df.rename(columns={"Age": "Student_Age"}, inplace=True) # Permanently changes the current data frame
```

```python
df.set_index("Name") # Changes default index 
```
